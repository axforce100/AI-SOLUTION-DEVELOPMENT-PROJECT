```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('bmarket.db')

# Read all data from the 'bank_marketing' table into a pandas DataFrame
df = pd.read_sql_query("SELECT * FROM bank_marketing;", conn)

# Close the connection
conn.close()

# Apply age cleaning as done in pJRZjM0o3ExG
df['Age'] = df['Age'].str.replace(' years', '', regex=False)
df['Age'] = pd.to_numeric(df['Age'])
df = df[df['Age'] < 100]

# Now, the 'Occupation' column should be present in df
print(df['Occupation'].unique())
```

**Reasoning**:
The previous attempts to identify rows with 'unknown' values failed because the DataFrame state was incorrect. Now that the DataFrame has been reloaded and its 'Occupation' column confirmed to contain 'unknown', I will re-execute the boolean masking to correctly identify all rows where 'unknown' exists in any column, as per the subtask instructions.



```python
rows_to_drop = df.isin(['unknown']).any(axis=1)
print(f"Number of rows with 'unknown' values: {rows_to_drop.sum()}")
```

**Reasoning**:
With `rows_to_drop` correctly identifying the rows containing 'unknown' values, the next step is to remove these rows from the DataFrame as per the task requirements. After removing the unwanted rows, I will display the head of the updated DataFrame to verify the removal and then print its information to show the new number of entries.



```python
df_cleaned = df[~rows_to_drop]
print("DataFrame head after removing 'unknown' rows:")
print(df_cleaned.head())
print("\nDataFrame info after removing 'unknown' rows:")
df_cleaned.info()
```
